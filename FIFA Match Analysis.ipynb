```python
sns.histplot(df["total_goals"], bins=10)

plt.title("Distribution of Total Goals per Match")
plt.ylabel("Number of Matches")
plt.xlabel("Total Counts")
plt.show()
```

```python

avg_goals= pd.DataFrame({
    "Team Types": ["home", "away"], 
    "Average goals":[
        df["home_score"].mean(),
        df["away_score"].mean()
    ]
})
```

```python
sns.barplot(x="Team Types", y="Average goals",
    data=avg_goals)  
plt.title("Average Goals: Home vs Away Teams")
plt.ylabel("Goals")
plt.show()
```

```python
df["home_win"] = df["home_score"]>df["away_score"]
```

```python
df["home_win"].value_counts()
```

```python
def get_result(row):
    if row["home_score"] > row["away_score"]:
        return "home win"
    elif row["home_score"] < row["away_score"]:
        return "away win"
    else: 
        return "Draw"
    
df["result"]= df.apply(get_result, axis=1)
```

```python
df["result"].value_counts()
```

```python
df["result"].value_counts(normalize=True)*100
```

```python
sns.countplot(x="result", data=df)

plt.title("Match Outcome Distribution")
plt.xlabel("Results")
plt.ylabel("Number of Matches")

plt.show()
```
